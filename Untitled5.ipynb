```python
#Importa a biblioteca Pandas
import pandas as pd
#Leitura do arquivo CSV de im√≥veis do RJ
pd.read_csv('ANALISE.csv')
```

```python
dados = pd.read_csv('ANALISE.csv', sep=';')
dados
```

```python
#Importar a biblioteca

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
reclamacoes =dados.groupby('MESES')["RECLAMACAO"].mean()

MESES = ['JANEIRO','FEVEREIRO','MAR√áO','ABRIL','MAIO','JUNHO','JULHO','AGOSTO','SETEMBRO','OUTUBRO','NOVEMBRO','DEZEMBRO']



#Cria as dimens√µes da figura do gr√°fico
fig, ax = plt.subplots(figsize = (17,17))
#Cor para o gr√°fico
cor  = ("#CC00FF", "#DF5DFF", "#EEBAFA", "#FFFFFF", "#BDEEF9","#97FFDC","#8BFFD8", "#75F1C8", "#6BFFCE",
"#51FFC5")

ax.bar(MESES,reclamacoes,width=0.6,color=cor,edgecolor="black")
ax.set_title("Reclama√ß√µes por mes", fontsize=20)
ax.set_ylabel("Reclama√ß√µes", fontsize=19)
ax.set_xlabel("Meses", fontsize=19)

for MESES, media in enumerate(reclamacoes):
  plt.text(x=MESES-.20,y=media+0.1,s=media,fontdict=dict(fontsize=20))
plt.tight_layout()
plt.show()




```
